Sin embargo, la categoria generada tiene un comportamiento que puede generar lecturas incorrectas. Por lo tanto, vamos a crear variables dummy para poder interpretar correctamente los valores categóricos.

```python
def dummy_variables(data, columnas):
    '''
    Convierte las columnas categoricas en dummys con valores booleans

    data: dataset a transformar
    columnas: iterable con los nombres de las columnas

    return: dataset transformado
    
    '''

    import pandas as pd
    data_columns = data[columnas]
    dummys = pd.get_dummies(data_columns, prefix= "Dummy", prefix_sep="_")
    insert_columns = dummys.columns.values.tolist()
    data[insert_columns] = dummys


    return pd.DataFrame(data)
```

```python
dummy = pd.get_dummies(data["Country"], prefix='dummy', prefix_sep="_")
dummy
```

```python
columns_new = dummy.columns.values.tolist()
data[columns_new] = dummy
data
```

```python
newData =  dummy_variables(data, "Country")
newData
```

```python
data_transformada = dummy_variables(data, "Country")
data_transformada.head()
```
