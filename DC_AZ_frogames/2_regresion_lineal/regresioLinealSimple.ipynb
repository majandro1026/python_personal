Las técnicas de Regresión son muy variadas, desde la regresión lineal hasta la SVR o la Regresión con Bosques Aleatorios.

En esta parte, vamos a entender e implementar los siguientes modelos de Regresión dentro del Machine Learning:

## Regresión lineal simple

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#data a trabajar
data = pd.read_csv('../material/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv')
data.head()
```

**Objetivo:** formular un modelo de regresión lineal que prediga el salario de una persona, dados sus años de experiencia, mediante la utilización de un algoritmo de regresión lineal simple.

### Definición: 
La regresión en estadistica es un proceso de estimar las relaciones existentes entre variables dependientes con respecto a una o mas variables independientes.

Para la regresión lineal simple, podemos determinar el siguiente modelo de ecuación lineal:

$$y = b_0 +b_1*x$$

Donde $b_1$ se conoce como la pendiente y $b_0$ el corte con el eje $y$.

```python
%matplotlib inline
plt.figure(figsize=(6,4))
plt.plot(data.YearsExperience, data.Salary, c = 'lightblue',marker = 'o')
plt.title('Años de experiencia Vs. Salario', fontsize = 20, color = 'gray')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')

plt.grid()
plt.show()
```

### método de los mínimos cuadrados

Para entender el modelo de regresión lineal, debemos interiorizar el concepto de minimos cuadrados, que no es mas que la diferencia que se encontrará entre un dato real $x_0$ y su respectiva distacia a un punto de la recta de la regresión $x_i$. La idea es minimizar esta diferencia para garantizar un mejor comportamiento del modelo.

$$ min \sum_{i}({y_i-\hat{y}_i})^2$$

#### Dividiendo el dataset en entrenamiento y testing

```python
x_train, x_test, y_train, y_test = train_test_split(data.YearsExperience, data.Salary, test_size=0.5, random_state=0)
x_train = np.array(x_train).reshape(1,-1)
y_train = np.array(y_train).reshape(1,-1)
x_test = np.array(x_test).reshape(1, -1)
```

```python
x_test.shape
```

Nota: el reshepe se debe hacer debido a la nueva version de la funcion **fit** del paquete sklearn.

#### Creando el modelo de regresión lineal

```python
regresion = LinearRegression()
regresion.fit(x_train, y_train)
```

#### Predecir el conjunto de test

```python
y_pred = regresion.predict(x_test)
```
